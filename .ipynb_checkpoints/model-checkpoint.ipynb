## Get the data

```python
path = Path('data')
```

```python
df = pd.read_csv(path/'histo/labels.csv')
df.head()
```

```python
df['id'] = df['id'].astype(str) + '.tif'
```

```python
tfms = get_transforms(flip_vert=True, max_lighting=0.1, max_zoom=1.05, max_warp=0.)
```

```python
data = ImageDataBunch.from_csv(path/'histo', folder='train', valid_pct=0.15, suffix='.tif', ds_tfms=tfms, size=48)
```

```python
data.show_batch(rows=3, figsize=(7, 8))
```

```python
len(data.train_ds), len(data.valid_ds)
```

```python
data.c
```

## Train model

```python
learn = create_cnn(data, models.resnet34, metrics=error_rate)
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
lr=1e-3
```

```python
learn.fit_one_cycle(10, slice(lr))
```

```python
learn.unfreeze()
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
learn.fit_one_cycle(10, slice(1e-3))
```

```python
learn.save('stage-one-again')
```

## Evaluate on test set

```python
test_data = (
    ImageItemList
        .from_df(df, path/'histo', folder="train")
        .random_split_by_pct(0.15)
        .label_from_df()
        .add_test_folder('test')
        .transform(tfms, size=96)
        .databunch()
        .normalize(imagenet_stats)
)
```

```python
learn.data = test_data
```

```python
len(learn.data.test_ds)
```

```python
predictions = learn.get_preds(ds_type=DatasetType.Test)
```

```python
len(predictions[0])
```

```python
_, preds = torch.max(predictions[0], 1)
```

```python
preds[999]
```

```python
def get_row(idx):
    pred = preds[idx].item()
    
    image_path = learn.data.test_ds.x.items[idx]
    image_name = re.search(r'([^/]+)$', f'{image_path}')[0].replace('.tif', '')
    
    return image_name, pred
```

```python
df = pd.DataFrame(columns=['id', 'label'])
for idx in range(len(predictions[0])):
    if idx % 5000 == 0:
        print(f"Completed: {idx}")
        
    image_name, label = get_row(idx)
    df.loc[idx] = [image_name, label]
```

```python
df.head()
```

```python
df.to_csv('sub.csv', index=False)
```
